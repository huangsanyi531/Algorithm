```
def BottomUpCutRod(p, n):
    r = [0]*(n+1)
    for i in range(1, n+1):
        if n == 0:
            return 0
        q =0
        p.append(0)  #由於沒有設定10以後長度的價格，每當計算長度超過10的問題時，會出現錯誤，所以將11以後長度的價格都設為0，讓程式跑成功
                #n輸入多少，後面就會跑出多少個0
        for j in range(1, i+1):
          q = max(q, p[j]+r[i-j]) #q為各種分割方式中，每一方式所得到的最大金額量，這邊取最大的數
          r[i] = q #r為一個紀錄每個長度可得到之最大金額的串列
          #print(r)
          #print(q)
    return r[n],r
n = int(input("請輸入數字:"))
p=[0,1,5,8,9,10,17,17,20,24,30] #價格
print(BottomUpCutRod(p, n))   #  (30, [0, 1, 5, 8, 10, 13, 17, 18, 22, 25, 30])
#print("p =",p)
```
