Selection sort (選擇排序) 2020/10/6

```
def Selction_Sort(datas):
    for i in range(len(datas)-1):          # (n-1)次
        min = i                  # n-2次
        for j in range(i+1, len(datas)):    # [n*(n-1)]/2次
            if datas[j] < datas[min]:
                min = j

        datas[i], datas[min] = datas[min], datas[i] # n-1次

        print(datas)                   #n-1次
        print('---'*20)                 #n-1次
    return datas
                        #BigO()=n^2

if __name__ == '__main__':
    import random
    a = list(range(16))
    random.shuffle(a)
    print(a)
    a = Selction_Sort(a)
    print(a)
```
